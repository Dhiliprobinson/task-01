```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_size = 1000
```

```python
np.random.seed(42)
```

```python
ages = np.random.randint(0, 101, population_size)
```

```python
genders = np.random.choice(['Male', 'Female'], population_size)
```

```python
data = {
    'Age': ages,
    'Gender': genders
}
df = pd.DataFrame(data)
```

```python
print("First few rows of the dataset:")
print(df.head())
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(10, 5))
gender_counts.plot(kind='bar', color='green', alpha=0.7)
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
